## Example Usage of Duration Class ##

### Setup ###

```python
import os
import sys

sys.path.append(os.path.dirname(os.getcwd()))
from io import StringIO

from IPython.display import display_html

from hec import Duration, Interval
```

### Generate a Table of Durations for all CWMS Intervals ###
Note that instantaneous durations are simultaneously BOP and EOP.

```python
buf = StringIO()
buf.write("<html><body><table>")
buf.write(
    '<tr><th style="text-align:left;">Interval</th><th>str(intvl)</th><th>intvl.name</th><th>BOP?</th>'
)
buf.write(
    '<td style="text-align:left;">Duration</th><th>str(dur)</th><th>dur.name</th>'
)
buf.write("<th>dur.is_bop</th><th>dur.is_eop</th></tr>")
for intvl in Interval.get_all_cwms():
    buf.write(
        f'<tr><td rowspan="2" style="text-align:left;">{repr(intvl)}</td><td rowspan="2">{str(intvl)}</td><td rowspan="2">{intvl.name}</td>'
    )
    for bop in (False, True):
        if bop:
            buf.write("<tr>")
        buf.write(f"<td>{bop}</td>")
        dur = Duration.for_interval(intvl, bop)
        buf.write(
            f'<td style="text-align:left;">{repr(dur)}</td><td>{str(dur)}</td><td>{dur.name}</td><td>{dur.is_bop}</td><td>{dur.is_eop}</td></tr>'
        )
buf.write("</table></body></html>")
html = buf.getvalue()
buf.close()
display_html(html, raw=True)  # type: ignore
```
